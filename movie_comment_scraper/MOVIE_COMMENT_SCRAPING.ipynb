## 1. 영화리뷰 텍스트 및 작성일자, 평점 수집하기

---

```python
from bs4 import BeautifulSoup 
import requests

# 수집할 영화리뷰 정보를 입력합니다.
PAGE_LIMIT = 10                         # 수집할 영화리뷰 페이지 개수
SAVE_FILE_PATH = "movie_comment.txt"    # 영화리뷰 저장 파일명
MOVIE_CODE = 194205                     # 영화 코드

f = open(SAVE_FILE_PATH, "w", encoding="utf-8")
for page in range(1, PAGE_LIMIT+1):
    print(page, end="\r")
    URL = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code="
    URL += str(MOVIE_CODE) + "&type=after&isActualPointWriteExecute=false"
    URL += "&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="
    URL = URL + str(page)
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")
    comment_list = soup.findAll("div", {'class': "score_reple",})     # 댓글/작성일자 추출
    score_list = soup.findAll("div",{'class': "star_score",})         # 평점 추출

    for comment, score in zip(comment_list, score_list):
        # 댓글
        p_span_tag_list = comment.find("p").findAll("span")
        if p_span_tag_list[0].text == "관람객":
            commtnet_text = p_span_tag_list[1].text.strip()
        else:
            commtnet_text = p_span_tag_list[0].text.strip()
        
        # 작성일자
        dt_em_tag_list = comment.find("dt").findAll("em")
        datetime = dt_em_tag_list[1].text.strip()
        
        # 평점
        em_tag = score.find("em")
        score = em_tag.text.strip()
        
        # 파일쓰기
        f.write(datetime+"\t"+ score+ "\t"+ commtnet_text + "\n")
        print(datetime+"\t"+ score+ "\t"+ commtnet_text)
    f.flush()
f.close()
```
